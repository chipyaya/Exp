```python
pos_words = [w.strip() for w in open('NTUSD_pos.txt')]
neg_words = [w.strip() for w in open('NTUSD_neg.txt')]
pos_set = set(pos_words)
neg_set = set(neg_words)
```

```python
for b in blogs:
    b['pos'] = 0
    b['neg'] = 0
for b in blogs:
    for x in chain.from_iterable(b['content']):
        if x in pos_set:
            b['pos'] += 1
        if x in neg_set:
            b['neg'] += 1
```

```python
for b in blogs:
    b['comment_pos'] = 0
    b['comment_neg'] = 0
for b in blogs:
    for x in chain.from_iterable(chain.from_iterable(b['comment'])):
        if x in pos_set:
            b['comment_pos'] += 1
        if x in neg_set:
            b['comment_neg'] += 1
```
